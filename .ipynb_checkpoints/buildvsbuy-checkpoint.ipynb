```python
# first thing we need to do is import our dependencies
# in this case it is pandas a dataframe tool 
# https://pandas.pydata.org/
import pandas as pd
```

```python
# we load our CSV into a python dataframe 
# There is a lot of rows so we need to remove memory constraints
# if this was a production level notebook you would probably do analysis by chunks through whats 
# called a file stream, load some rows, do analysis/cleaning, load the next set
gc_spending_by_year = pd.read_csv(
    "./data/GC_Spending_by_year.csv",
    index_col = 0,
    low_memory=False
)
```

```python
# gives you some overall statistics of your dataset 
gc_spending_by_year.describe()
```

```python
# we have over a million rows ! 
gc_spending_by_year.shape[0]
```

```python
# so we want to make all lower case so it's cleaner to filter on
gc_spending_by_year["description"] = gc_spending_by_year["description"].str.lower()
gc_spending_by_year["extra_description"] = gc_spending_by_year["extra_description"].str.lower()
```

```python
# here is our first 5 rows 
gc_spending_by_year.head()
```

```python
# we are combining the two description rows into one column to make it simpler to filter 
gc_spending_by_year["combined_description"] = gc_spending_by_year[
    "description"
].fillna('') + " "  + gc_spending_by_year[
    "extra_description"
].fillna('')
```

```python
gc_spending_by_year.head()
```

```python
# filtering on combined_description column, & and ~ are bitwise operators, you can thing of them as (and) and (not)
gc_spending_by_year_software_projects = gc_spending_by_year[
    gc_spending_by_year["combined_description"].str.contains("software") &
    ~ gc_spending_by_year["combined_description"].str.contains("maintenance")
]
```

```python
# our rows now, much reduced !
gc_spending_by_year_software_projects.shape[0]
```

```python
# saving our reduced and filtered dataset 
gc_spending_by_year_software_projects.to_csv(
    "./data/GC_Spending_by_year_software.csv",
    encoding="utf-8"
)
```
